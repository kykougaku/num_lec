# 実際にhorner-newton法で解いてみる

```python
# horner - newton

def horner_newton(funca, x0, eps=1e-3):
    alplhalist = []

    xk = x0
    pna = funca
    for i in range(len(funca) - 2):
        while abs(np.polyval(pna,xk)) > eps:
            pnxk = np.polyval(pna,xk)
            pnm1xk = np.polyval(np.polydiv(pna, np.array([1.0, -xk], np.float128))[0], xk)
            xk = xk - (pnxk / pnm1xk)

        alplhalist.append(xk)
        pna = np.polydiv(pna, np.array([1.0, -xk], np.float128))[0]
    return alplhalist
```

```python
print(horner_newton(funca, 5.0))
```
