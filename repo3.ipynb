```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func (x):
    return (x-1)*np.log(x) - 1
```

```python
def dfunc (x):
    return np.log(x) + (x-1)/x
```

## 関数の概形

```python
xlist = np.linspace(1e-10, 10, 1000)
ylist = func(xlist)
plt.plot(xlist, ylist)
plt.grid()
plt.title('f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

xlist = np.linspace(0.1, 10, 1000)
ylist = dfunc(xlist)
plt.plot(xlist, ylist)
plt.grid()
plt.title('f\'(x)')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.show()
```

## 二分法とニュートン法の比較

```python
def bisection (a, b, func, eps=1e-6, delta=1e-10):
    if func(a)*func(b) > 0:
        print("No root in this interval")
        return None
    resuduallist = []
    iterlist = []
    iter = 0
    mid = (a+b)/2
    while (b-a) > eps or np.abs(func(mid)) > delta:
        mid = (a+b)/2
        iter += 1
        resuduallist.append(np.abs(func(mid)))
        iterlist.append(iter)
        if func(mid) == 0:
            return mid
        elif func(a)*func(mid) < 0:
            b = mid
        else:
            a = mid
    return mid, resuduallist, iterlist
```

```python
def newton (a, b, func, dfunc, eps=1e-10, delta=1e-15):
    if func(a)*func(b) > 0:
        print("No root in this interval")
        return None
    resuduallist = []
    iterlist = []
    iter = 0
    x = (a+b)/2
    while np.abs(func(x)) > delta:
        iter += 1
        resuduallist.append(np.abs(func(x)))
        iterlist.append(iter)
        x = x - func(x)/dfunc(x)
    return x, resuduallist, iterlist
```

```python
mid, resuduallist, iterlist = bisection(1, 10, func)
plt.plot(iterlist, resuduallist)
plt.yscale('log')
plt.grid()
plt.title('Bisection')
plt.xlabel('Iteration')
plt.ylabel('Residual')
plt.show()
```

```python
x, resuduallist, iterlist = newton(1, 10, func, dfunc)
plt.plot(iterlist, resuduallist)
plt.yscale('log')
plt.grid()
plt.xlabel('Iteration')
plt.ylabel('Residual')
plt.title('Newton')
plt.show()
```
